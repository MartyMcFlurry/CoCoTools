## Utility functions

```python
def figsave(name, fig=None, ext='pdf'):
    if fig is None:
        fig = plt.gcf()
    fig.savefig('%s/%s.%s' % (figdir, name, ext))
```

```python
def graph_view(graph, projection, size=None, only_nodes=None, node_alpha=None, edge_alpha=None, 
               title=None):
    """Simple utility to draw the various projections.

    WARNING: this function relies on the presence of globals named 
    aspect, colors and gnorm!!!
    """
    from cocotools import nxdraw; reload(nxdraw)
    
    if size is None:
        size = dict(axial=6, sagittal=12, coronal=6)[projection]
        
    f, ax = plt.subplots(figsize=(size, size/aspect[projection]))
    if title is None:
        title = "%s Projection" % projection.title()
        
    ax.set_title(title)
    
    # Smart defaults for node/edge transpanrency: for full graphs, it's better
    # to make the edges faint and the nodes darker
    if only_nodes is None: n_alpha=0.7; e_alpha=0.1
    else: n_alpha=0.9; e_alpha=0.25
    # User can still override by giving values
    node_alpha = n_alpha if node_alpha is None else node_alpha
    edge_alpha = e_alpha if edge_alpha is None else edge_alpha
        
    nxdraw.draw_network(graph, pos=positions[projection], ax=ax, node_color=colors,
                        only_nodes=only_nodes, node_alpha=node_alpha, edge_alpha=edge_alpha)
    
    if projection=='axial':
        ax.set_xlim(-3,30)
    return ax
```

## Loading global data and constants needed by all plots

This cell defines the map and kind of graph we want to load for making plots

```python
# Define map type to load here:
#graph_map = 'r00'
graph_map = 'pht00'
#graph_map = 'fv91'

# Define graph type:
graph_type = 'endg_mort'
#graph_type = 'native'
```

```python
# The rest below shouldn't need any further tweaking

# Construct graph file path and atlas spec
graph_name = os.path.join('..', 'graphs', '%s_%s.pck' % (graph_map, graph_type))

import utils; reload(utils)

# Load standard labels
atlas_kind = getattr(utils, graph_map)

atlas = utils.make_atlas(atlas_kind)
# Compute quantities needed from the atlas
labels, positions, aspect, colors = utils.make_projections(atlas)

# Load the graph, both as it comes out and normalized to only have
# nodes for which we have coordinates
graph, gnorm = utils.load_graphs(graph_name, atlas)

# Print some basic summary of the loaded graph for sanity check
subcort={'pht00':['AA','AHIMC', 'BM', 'BL', 'CE', 'CM' 'LA', 'ME'], 'r00' :['AMYG', 'HC']}

graph.remove_nodes_from(subcort[graph_map])

# Print some basic summary of the loaded graph for sanity check
print 'Nodes in original graph:'
print sorted(graph.nodes())
print '\nNodes for which we have coordinates:'
print sorted(gnorm.nodes())
discarded = set(graph.nodes()) - set(gnorm.nodes())
print '\nDiscarded nodes:', len(discarded),' total\n', sorted(discarded)
```
